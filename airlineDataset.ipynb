```python
col = "Arrival Delay in Minutes"

# Calculate the 25th percentile (Q1) and 90th percentile (Q3) for the selected column
Q1 = df_copy[col].quantile(0.25)
Q3 = df_copy[col].quantile(0.90)

# Compute the interquartile range (IQR) using Q1 and Q3
IQR = Q3 - Q1

# Define the upper bound using an extended IQR multiplier (k=2.0)
k = 2.0
upper_bound_iqr = Q3 + k * IQR

# Filter out values considered outliers (greater than upper bound)
outliers_iqr = df_copy[df_copy[col] > upper_bound_iqr]

# Print the upper bound and number of detected outliers
print(f"Upper bound: {upper_bound_iqr:.2f}")
print(f"Number of Outlier: {len(outliers_iqr)}")
```

```python
from scipy.stats import zscore

col = "Flight Distance"

# Calculate Z-scores for the column (standardized values)
z_scores = zscore(df_copy[col])

# Set the Z-score threshold for identifying outliers
threshold = 2.5

# Identify all outliers where absolute Z-score > threshold
outliers_z = df_copy[(z_scores < -threshold) | (z_scores > threshold)]

# Subset: flights with unusually short distances
outliers_small = df_copy[z_scores < -threshold]

# Subset: flights with unusually long distances
outliers_large = df_copy[z_scores > threshold]

# Print summary of Z-score based outlier detection
print(f"\n--- {col} ---")
print(f" Outlier threshold: |Z| > {threshold}")
print(f" Total number of outliers: {len(outliers_z)}")
print(f" Very short flights (Z < -{threshold}): {len(outliers_small)}")
print(f" Very long flights (Z > {threshold}): {len(outliers_large)}")

# Additionally, identify manually defined short flights under 150 km
short_outliers_manual = df_copy[df_copy["Flight Distance"] < 150]
print(f"Number of short flights (<150 km): {len(short_outliers_manual)}")
```

```python
col = "Departure Delay in Minutes"

# Calculate the 25th (Q1) and 90th (Q3) percentiles
Q1 = df_copy[col].quantile(0.25)
Q3 = df_copy[col].quantile(0.90)

# Compute the interquartile range (IQR)
IQR = Q3 - Q1

# Set the IQR multiplier
k = 2.0
upper_bound_iqr = Q3 + k * IQR

# Identify values that exceed the upper bound (considered outliers)
outliers_iqr = df_copy[df_copy[col] > upper_bound_iqr]

print(f"Upper bound: {upper_bound_iqr:.2f}")
print(f"Number of outliers: {len(outliers_iqr)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

col = "Arrival Delay in Minutes"

# Calculate the 25th percentile (Q1) and 90th percentile (Q3)
Q1 = df_copy[col].quantile(0.25)
Q3 = df_copy[col].quantile(0.90)

# Compute the interquartile range (IQR)
IQR = Q3 - Q1

# Calculate extended lower and upper bounds
lower = Q1 - 2 * IQR
upper = Q3 + 2 * IQR

# Create histogram with KDE
plt.figure(figsize=(10, 5))
sns.histplot(df_copy[col], bins=50, kde=True)

plt.axvline(upper, color='green', linestyle='--', label=f'Upper Bound ({upper:.1f})')

plt.xlim(0, 300)

plt.title("Arrival Delay in Minutes - Histogram with IQR Upper Bound")
plt.xlabel("Arrival Delay in Minutes")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import zscore

col = "Flight Distance"

# Calculate Z-scores for the column
z_scores = zscore(df_copy[col])
threshold = 2.5

# Filter out outliers:
# Keep values where |Z| <= 2.5 AND Flight Distance >= 150 km
df_flight_clean = df_copy[
    (abs(z_scores) <= threshold) & (df_copy[col] >= 150)
]

print(f"Cleaned Data: {df_flight_clean.shape}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create histogram for the cleaned "Flight Distance" column
plt.figure(figsize=(10, 5))
sns.histplot(df_flight_clean[col], bins=50, kde=True)

plt.title("Flight Distance - Histogram (Outliers Removed)")
plt.xlabel("Flight Distance (km)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Calculate the number of rows before and after outlier removal
total = df_copy.shape[0]
remaining = df_flight_clean.shape[0]
removed = total - remaining

print(f"Total data: {total}")
print(f"Cleaned data: {remaining}")
print(f"Number of outliers removed: {removed}")
```

```python
from scipy.stats import zscore
import matplotlib.pyplot as plt
import seaborn as sns

col = "Arrival Delay in Minutes"
threshold = 2.5

# Calculate Z-scores
z_scores_arrival = zscore(df_copy[col])

# Filter non-outliers
df_arrival_clean = df_copy[abs(z_scores_arrival) <= threshold]

# Count rows
total_rows = df_copy.shape[0]
remaining_rows = df_arrival_clean.shape[0]
removed_rows = total_rows - remaining_rows

print(f"\n--- {col} ---")
print(f"Total rows: {total_rows}")
print(f"Remaining after outlier removal: {remaining_rows}")
print(f"Outliers removed: {removed_rows}")


plt.figure(figsize=(10, 5))
sns.histplot(df_arrival_clean[col], bins=50, kde=True)
plt.title(f"{col} - Histogram (After Outlier Removal)")
plt.xlabel(col)
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
